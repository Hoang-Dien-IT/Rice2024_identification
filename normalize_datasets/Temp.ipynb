Import thư viện


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv('./Rice2024.csv')
print(df.head())
```

```python
# Xóa cột ‘Id’ và ‘Nickname’ Xóa cột ‘Id’ và ‘Nickname’
df = df.drop(['Id', 'Nickname'], axis=1)
print(df.head())
```

```python
# Xóa các hàng có dữ liệu thiếu
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
plt.figure(figsize=(10, 6))
missing_values.plot(kind='bar', color='orange')
plt.title('Biểu đồ số lượng dữ liệu thiếu của từng thuộc tính')
plt.ylabel('Number of Missing Values')
plt.show()
print("Đã xóa các hàng bị thiếu dữ liệu.")
```

```python
# xóa hàng 429,,,,0830078125
print("Xóa hàng Perimeter = 429,,,,0830078125")
df = df[df['Perimeter'] != '429,,,,0830078125']
```

```python
# Cột 'Area'
# Vẽ boxplot cho một thuộc tính (cột) 'A'
# Vẽ boxplot cho từng cột số
for column in df.columns:
    if column != 'Id' and column != 'Class':  # Loại bỏ cột 'id' nếu cột này tồn tại
        plt.figure(figsize=(8, 6))
        sns.boxplot(data=df[column])  # Vẽ boxplot cho thuộc tính cụ thể
        plt.title(
            f'Biểu đồ Boxplot kiểm tra ngoại lệ cho thuộc tính {column}', fontsize=16)
        plt.xlabel(f'{column}', fontsize=12)
        plt.ylabel('Giá trị', fontsize=12)
        plt.show()
        
```

```python
# Xóa hàng dữ liệu nhiễu Major_Axis_Length
print("Tìm Major_Axis_Length nhiễu: ", df['Major_Axis_Length'].max())
print("Đã xóa dữ liệu Major_Axis_Length nhiễu")
df_droped = df.drop(df['Major_Axis_Length'].idxmax())

print("Tìm Major_Axis_Length nhiễu: ", df_droped['Major_Axis_Length'].max())
print("Đã xóa dữ liệu Major_Axis_Length nhiễu")
df_droped = df_droped.drop(df_droped['Major_Axis_Length'].idxmax())

# Xóa hàng dữ liệu nhiễu Minor_Axis_Length
print("Tìm Minor_Axis_Length nhiễu: ", df_droped['Minor_Axis_Length'].max())
print("Đã xáo dữ liệu Minor_Axis_Length nhiễu")
df_droped = df_droped.drop(df_droped['Minor_Axis_Length'].idxmax())

for column in df_droped.columns:
    if column != 'Id' and column != 'Class':  # Loại bỏ cột 'id' nếu cột này tồn tại
        plt.figure(figsize=(8, 6))
        sns.boxplot(data=df_droped[column])  # Vẽ boxplot cho thuộc tính cụ thể
        plt.title(
            f'Biểu đồ Boxplot kiểm tra ngoại lệ cho thuộc tính {column}', fontsize=16)
        plt.xlabel(f'{column}', fontsize=12)
        plt.ylabel('Giá trị', fontsize=12)
        plt.show()
```

```python
# Countplot - Biểu đồ đếm số lượng mẫu giữa các lớp
plt.figure(figsize=(8, 6))
# Sử dụng bảng màu 'viridis' cho màu sắc khác nhau
sns.countplot(x='Class', data=df_droped, palette='viridis')
plt.title('Countplot of Rice Grain Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()


print("c -> C, o -> O")
df_droped['Class'] = df_droped['Class'].replace({'c': 'C', 'o': 'O'})

# Thực hiện Label Encoding cho các cột 'Class'
df_droped.loc[:, 'Class'] = df_droped['Class'].map({'C': 0, 'O': 1})
print("Label Encoding cho cột 'Class': C -> 0, O -> 1")


# Countplot - Biểu đồ đếm số lượng mẫu giữa các lớp
plt.figure(figsize=(8, 6))
# Sử dụng bảng màu 'viridis' cho màu sắc khác nhau
sns.countplot(x='Class', data=df_droped, palette='viridis')
plt.title('Countplot of Rice Grain Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Scale min-max
print("Dữ liệu trước khi scale:")
print(df_droped.head())

scaler = MinMaxScaler(feature_range=(0, 1))
# Chọn các cột số để chuẩn hóa
columns_to_scale = ['Area', 'Perimeter',
                    'Major_Axis_Length', 'Minor_Axis_Length', 'Eccentricity', 'Extent', 'Convex_Area']
# Tạo một bản sao của DataFrame để chuẩn hóa
df_scaled = df_droped.copy()
# Áp dụng chuẩn hóa Min-Max
df_scaled[columns_to_scale] = scaler.fit_transform(
    df_scaled[columns_to_scale])

print("Dữ liệu sau khi scale:")
print(df_scaled.head())
```
